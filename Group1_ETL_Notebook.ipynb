```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Filepath for facility data file
facility_csv = "PA_facility_pollution.csv"

# Read the csv file
facility_data = pd.read_csv(facility_csv)

# Display the data table for preview
facility_data.head()
```

```python
facility_clean = facility_data[["FacilityName", "Latitude", "Longitude", "LocationAddress", "City", "State", "ZIP", "County", "Industry_Clean", "TRI_Air_Emissions_10_in_lbs", "TRI_Air_Emissions_11_in_lbs", "TRI_Air_Emissions_12_in_lbs", "TRI_Air_Emissions_13_in_lbs", "TRI_Air_Emissions_14_in_lbs", "GHG_Direct_Emissions_10_in_metric_tons", "GHG_Direct_Emissions_11_in_metric_tons", "GHG_Direct_Emissions_12_in_metric_tons", "GHG_Direct_Emissions_13_in_metric_tons", "GHG_Direct_Emissions_14_in_metric_tons"]]
facility_clean.head()
```

```python
# Filepath for cancer data file
cancer_csv = "all_cancer_incidence.csv"

# Read the csv file
cancer_data = pd.read_csv(cancer_csv)

# Display the data table for preview
cancer_data.head()
```

```python
# cancer_data = cancer_data.set_index(cancer_data['County'])
cancer_data = cancer_data.drop(columns = ["RateRatio", "Cancer_Site", "Sex", "Race_Ethnicity", "Age", "SignificanceResult", "County_Code"])
cancer_data
```

```python
#Connect to local database 
rds_connection_string = "postgres:postgres@localhost:5432/group1_etl_project"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Confirming table names are present in SQL database
engine.table_names()
```

```python
#Uploading county, cancer, and facility level data to Postgres

facility_clean.to_sql(name='Facilities', con=engine, if_exists='append', index=False)
cancer_data.to_sql(name='Cancer_Data', con=engine, if_exists='append', index=False)
```

```python
# Query database
pd.read_sql_query('select * from "Cancer_Data"', con=engine).head()

```
